```python
# matplotlib_exercises
import matplotlib.pyplot as plt


# 1. Plot y = x**2 - x + 2
# You'll need to write the code that generates the x and y points.
# Add an anotation for the point 0, 0, the origin.

x = 12

exes = [x + 1 for x in range(x)]
whys = [x**2 - x + 2 for x in range(x)]

plt.plot(exes, whys)
plt.text(0, 0, ('Origin'), c = 'red')
plt.show()


# 2. Create and label 4 separate charts for:
# • y = √x
# • y = x**3
# • y = 2**x
# • y = 1/(x + 1)

x = [4, 8, 12, 16]
y = [x**(1/2) for x in x]

plt.plot(x, y)
plt.title('$y = √x$')
plt.show()

x = [0, .1, .5, 1, 1.5, 2]
y = [x**3 for x in x]

plt.plot(x, y)
plt.title('$x^3$')
plt.show()

x = [-3, -2, -1, -.5, -.1,0, .1, .5, 1, 1.5, 2, 2.5, 3 ]
y = [2**x for x in x]
plt.plot(x,y)
plt.title('$2^x$')
plt.show()

x = [0, 3, 6, 9, 12, 15]
y = [1/(x + 1) for x in x]

plt.plot(x, y)
plt.title('$1/(x + 1)$')
plt.show()

# 3. Combine the figures you created in the last step into one large figure with 4 subplots.
x = [4, 8, 12, 16]
y = [x**(1/2) for x in x]

plt.subplot(3, 2, 1)
plt.plot(x, y)
plt.title('$√x$')

x = [0, .1, .5, 1, 1.5, 2]
y = [x**3 for x in x]

plt.subplot(3, 2, 2)
plt.plot(x, y)
plt.title('$x^3$')

x = [-3, -2, -1, -.5, -.1,0, .1, .5, 1, 1.5, 2, 2.5, 3 ]
y = [2**x for x in x]
plt.subplot(3, 2, 3)
plt.plot(x,y)
plt.title('$2^x$')

x = [0, 3, 6, 9, 12, 15]
y = [1/(x + 1) for x in x]
plt.subplot(3, 2, 4)
plt.plot(x, y)
plt.title('$1/(x + 1)$')
```

```python
# 4. Combine the figures you created in the last step into one figure where each
# of the 4 equations has a different color for the points. Be sure to include a 
# legend and an appropriate title for the figure.
x = [4, 8, 12, 16]
y = [x**(1/2) for x in x]

plt.scatter(x, y, c = 'r')
plt.plot(x, y, label='$√x$')

x = [0, .1, .5, 1, 1.5, 2]
y = [x**3 for x in x]
plt.scatter(x, y, c = 'm')
plt.plot(x, y, label='$x^3$')

x = [-3, -2, -1, -.5, -.1,0, .1, .5, 1, 1.5, 2, 2.5, 3 ]
y = [2**x for x in x]
plt.scatter(x, y, c = 'k')
plt.plot(x,y, label='$2^x$')

x = [0, 3, 6, 9, 12, 15]
y = [1/(x + 1) for x in x]
plt.scatter(x, y)
plt.plot(x, y, label='$1/(x + 1)$')

plt.legend(loc='upper right')
plt.title('$y$ Comparison Charts')
```
